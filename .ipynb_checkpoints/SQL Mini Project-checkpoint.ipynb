```python
import pandas as pd
import os
import sqlite3

conn = sqlite3.connect('country_club.db')
cursor = conn.cursor()
```

```python
#print(os.listdir())
#SQLdata = pd.read_csv('SQLExport.csv') 
Bookings = pd.read_csv('country_club_table_Bookings.csv')
Facilities = pd.read_csv('country_club_table_Facilities.csv')
Members = pd.read_csv('country_club_table_Members.csv')
```

```python
Bookings.to_sql("Bookings", conn, if_exists="replace", index=False)
Facilities.to_sql("Facilities", conn, if_exists="replace", index=False)
Members.to_sql("Members", conn, if_exists="replace", index=False)

conn.commit()
```

```python
#PART 2: SQLite

# Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members! */

Q10 = """SELECT 
    f.name AS facility_name,
    SUM(CASE 
            WHEN b.memid = 0 THEN f.guestcost * b.slots  
            ELSE f.membercost * b.slots  
        END) AS total_revenue
FROM Bookings b
JOIN Facilities f 
    ON b.facid = f.facid
WHERE (b.memid != 0)
GROUP BY f.name
HAVING total_revenue < 1000  
ORDER BY total_revenue DESC;"""

pd.read_sql(Q10, conn)
```

```python
# Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */

Q11 = """SELECT 
    m1.surname, 
    m1.firstname, 
    m2.surname AS recommender_surname, 
    m2.firstname AS recommender_firstname
FROM Members m1
JOIN Members m2 ON m1.recommendedby = m2.memid
ORDER BY m1.surname, m1.firstname;"""

pd.read_sql(Q11, conn)
```

```python
# Q12: Find the facilities with their usage by member, but not guests */

Q12 = """
SELECT 
    f.name AS facility_name,
    COUNT(b.bookid) AS usage_count
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
WHERE b.memid <> 0 
GROUP BY f.name

;"""

pd.read_sql(Q12, conn)
```

```python
# Q13: Find the facilities usage by month, but not guests */

Q13 = """
SELECT 
    f.name AS facility_name,
    strftime('%m', b.starttime) AS month, 
    COUNT(b.bookid) AS usage_count
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
WHERE b.memid <> 0  
GROUP BY f.name, month
ORDER BY month;

"""

pd.read_sql(Q13, conn)
```
