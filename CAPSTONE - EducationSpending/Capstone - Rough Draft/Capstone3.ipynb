## 2012

```python
url = "https://nces.ed.gov/programs/digest/d14/tables/dt14_226.40.asp"

tables = pd.read_html(url, header=0)

df = tables[3]

thistitle = 'sat_2012'
```

```python
df['State'] = df['State'].astype(str).str.rstrip('34')

df = df[df['State'].notna() & (df['State'].str.strip() != '')]

rows = df['State'].str.strip().isin(['District of', 'Colombia'])
if rows.sum() == 2:
    combined_row = df.loc[rows].drop(columns=['State']).apply(pd.to_numeric, errors='coerce').sum()
    df = df[~rows] 
    df.loc[len(df)] = ['District of Columbia'] + combined_row.tolist()

df = df.iloc[:, :-2]
df = df.loc[:, df.columns.str.startswith('2012') | df.columns.str.startswith('State')]
df.index = [thistitle if i == df.index[0] else i for i in df.index]

df.iloc[0] = ["State", "Reading", "Mathematics", "Writing"]
df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop=True)

df = df.iloc[1:].reset_index(drop=True)

df.set_index(df.columns[0], inplace=True)

df = df[df.index != 'nan']

df = df.drop("District of", errors='ignore')
df.rename(index={"Columbia": "District of Columbia"}, inplace=True)
```

```python
df[df.columns] = df[df.columns].astype(int)

sat_2012 = df

print(sat_2012)
```

```python
national_avg = df_2012.iloc[0]  #National Average is the first row value
state_data = df_2012.iloc[1:]   #And since we don't want to count that with everything else...

subjects = ['Composite score', 'English score', 'Mathematics score', 'Reading score', 'Science score']

for subject in subjects:
    plt.figure(figsize=(16, 6))
    state_data[subject].plot(kind='bar', color='skyblue', edgecolor='black')
    plt.axhline(y=national_avg[subject], color='red', linestyle='--', label='National Avg')
    plt.title(f'{subject} by State (vs National Average)', fontsize=16)
    plt.ylabel('Score')
    plt.xlabel('State')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
for subject in subjects:
    # Sorting the states by the subject scores
    sorted_states = state_data[subject].sort_values(ascending=False)
    top5 = sorted_states.head(5)
    bottom5 = sorted_states.tail(5)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))
    
    # Plot for the top 5
    top5.plot(kind='bar', ax=ax1, color='lightgreen', edgecolor='black')
    ax1.axhline(national_avg[subject], color='red', linestyle='--', label='National Avg')
    ax1.set_title(f'Top 5 States - {subject}', fontsize=14)
    ax1.set_ylabel('Score')
    ax1.set_ylim([min(bottom5.min(), national_avg[subject]) - 1, max(top5.max(), national_avg[subject]) + 1])
    ax1.legend()
    
    # Plot for the top 5
    bottom5.plot(kind='bar', ax=ax2, color='orange', edgecolor='black')
    ax2.axhline(national_avg[subject], color='red', linestyle='--', label='National Avg')
    ax2.set_title(f'Bottom 5 States - {subject}', fontsize=14)
    ax2.set_ylabel('Score')
    ax2.set_ylim([min(bottom5.min(), national_avg[subject]) - 1, max(top5.max(), national_avg[subject]) + 1])
    ax2.legend()
    
    ax1.tick_params(axis='x', rotation=45)
    ax2.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
```
