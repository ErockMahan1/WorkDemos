```python
category_trend = data.groupby(['published', 'Category']).size().unstack(fill_value=0)

category_trend.plot(kind='line', marker='o')

plt.title('Video Count per Category by Year', fontsize=14)
plt.xlabel('Published Year')
plt.ylabel('# of Videos')
plt.legend(title='Category', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#While 20 years ago there were different popular items, since around 2019 we've seen a steadily increasing number of music videos
#The only reason the value drops so much in 2025 is that we're not even halfway through the year yet.
#Assuming a steady release of video content throughout the year, that does mean there is a downward trend.
#But it is still exponentially higher than the other category options
```

```python
#Here is a plot of the percentages as well:

category_counts = data.groupby(['published', 'Category']).size().unstack(fill_value=0)

category_pct = category_counts.div(category_counts.sum(axis=1), axis=0) * 100


plt.figure(figsize=(12, 6))
category_pct.plot(kind='line', marker='o', figsize=(12, 6))

plt.title('% of Videos per Category by Year', fontsize=14)
plt.xlabel('Published Year')
plt.ylabel('Category Share (%)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#In terms of percentage, there were quite a few Gaming videos produced that dominated the market
#That would have been during the COVID 19 pandemic, and there has been a major drop since the quarantine period ended.
#Music still dominates the market.
```

```python
#Now let's try to guess where things are going in the future using linear regression...
#It's getting crowded, so I'm just going to focus on these categories:

selected_categories = [
    'People & Blogs',
    'Gaming',
    'Music'
]

#These seem to be the most popular at the moment, based on previous data

filtered_data = data[data['Category'].isin(selected_categories)]

category_trend = filtered_data.groupby(['published', 'Category']).size().unstack(fill_value=0)

last_year = category_trend.index.max()
all_years = np.arange(category_trend.index.min(), last_year + 11).reshape(-1, 1)

for category in category_trend.columns:
    years = category_trend.index.values.reshape(-1, 1)
    counts = category_trend[category].values

    model = LinearRegression()
    model.fit(years, counts)

    predictions = model.predict(all_years)

    plt.plot(years, counts, label=f'{category} (actual)')

    plt.plot(all_years, predictions, linestyle='dashed', label=f'{category} (predicted)', alpha=0.7)

plt.title('Actual and Predicted Video Counts per Category (Next 10 Years)')
plt.xlabel('Year')
plt.ylabel('# of Videos')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
#Through linear regression, the data strongly suggest that 'Music' video categories are the way of the future!
#Alas, most others are dying out, but we can watch the data easily enough and see where things go!
```
